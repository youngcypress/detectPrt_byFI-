```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']#设置可以显示中文
plt.rcParams['axes.unicode_minus'] = False #正常显示负号


color_set = [
    "#be002e", "#c0c0c0", "#FFA500", "#72001c", "#1c335b", "#ffaa96",
    "#dc3400", "#5b9bd5", "#f32e00", "#01b8aa", "#f2c80f", "#016e66"
]

df_info = pd.read_csv("参考指数收益率.csv", index_col='date')
df_info
```

```python

a = df_info.corr()
plt.subplots(figsize=(18, 18), dpi=120)
mask1 = np.triu(np.ones_like(a, dtype=bool))
mask2 = np.abs(a) <= 0.5
mask  = mask1 | mask2
sns.heatmap(a, annot=True, vmax=1, vmin=0.5, square=True, cmap='Blues')
plt.savefig('仓位相关系数矩阵热力图.png')
plt.show()
```
